# Matplotlib
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
# Import
import numpy as np
import matplotlib.pyplot as plt
```

# Figure and Axes
The figure keeps track of all the child axes and even nested subfigures. The axes are a group of 'special' artists such as titles, figure legends, color bars, etc. Multiple axes are possible in a figure.

```python
# Figure
fig, ax = plt.subplots(figsize=(5,4))
```

```python
# Figure and Axes
fig, ax = plt.subplots(figsize=(5,4))
ax.plot(np.arange(1, 5), np.random.rand(4) * 10)
```

```python
# Subplots
fig, axs = plt.subplots(ncols=2, nrows=2)
for r in range(2):
    for c in range(2):
        axs[r, c].plot(np.arange(1, 5), np.random.rand(4) * 10)
```

```python
# Mosaic
fig, axd = plt.subplot_mosaic([['upper left', 'right'],
                               ['lower left', 'right']])

for k in axd.keys():
    axd[k].plot(np.arange(1, 5), np.random.rand(4) * 10)
```

# Explanation
The data and graph can be explained in different ways, such as by titling, labeling, texting, annotating, scaling, combining, etc.

```python
# Title, xlabel, and ylabel
fig, ax = plt.subplots(figsize=(5,4))
ax.plot(np.arange(1, 5), np.random.rand(4) * 10)
ax.set_title('Graph')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
```

```python
# Legend, text, and annotate
fig, ax = plt.subplots(figsize=(5,4))
ax.plot(np.arange(1, 11), np.arange(1, 11)**2, label=r'$y=x^2$')
ax.text(8, 90, r'$y=x^2$')
ax.annotate(r'$y=x^2$', xy=(5, 25), xytext=(6, 10),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.legend()
```

```python
# Scale
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
data1 = np.random.randn(100)
xdata1 = np.arange(len(data1))  # make an ordinal for this
data2 = 10**data1

axs[0].plot(xdata1, data2)
axs[1].set_yscale('log')
axs[1].plot(xdata1, data2)
```

```python
# Twin
fig, ax1 = plt.subplots(figsize=(5, 4))
x = np.arange(1,11)

ax1.plot(x, x**2, color="blue")
ax1.set_ylabel(r"area $(m^2)$", color="blue")
    
ax2 = ax1.twinx()
ax2.plot(x, x**3, color="red")
ax2.set_ylabel(r"volume $(m^3)$", color="red")
```

# Style
The data can be shown in a better way with styling. Only the basic plot is included here. Similar or different styles can be used for other plot types.

```python
# Plot style
x = np.arange(1, 11)

fig, ax = plt.subplots()

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10])

ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
```

```python
# Plot style
x, y, z, t = np.random.randn(4, 100)
fig, ax = plt.subplots()
ax.plot(x, 'o')
ax.plot(y, 'd')
ax.plot(z, 'v')
ax.plot(t, 's')
```

# Examples
There are many plot types for different kinds of interests. Only the basic plots that are related to data science are included here.

### Basic Plots

```python
# Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(5,4))
ax.plot(x, y)
```

```python
# Scatter
x = np.random.rand(50)
y = np.random.rand(50)

fig, ax = plt.subplots(figsize=(5,4))
ax.scatter(x, y)
```

```python
# Bar
x = ['A', 'B', 'C', 'D', 'E']
y = [3, 7, 2, 5, 8]

fig, ax = plt.subplots(figsize=(5,4))
ax.bar(x, y)
```

### Statistics Plots

```python
# Hist
x = np.random.randn(1000)

fig, ax = plt.subplots(figsize=(5,4))
ax.hist(x)
```

```python
# Boxplot
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)

data = [x, y, z]

fig, ax = plt.subplots(figsize=(5,4))
ax.boxplot(data)
```

```python
# Pie
x = [25, 30, 20, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']

fig, ax = plt.subplots(figsize=(5,4))
ax.pie(x, labels=labels)
```
