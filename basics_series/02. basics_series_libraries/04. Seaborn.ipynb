# Seaborn
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
# Import
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data
These are the built-in data sets that Seaborn provides.

```python
# Data
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
# Data
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Data
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
# Data
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Data
flights_long = sns.load_dataset('flights')
flights_long
```

# Relational Plots
Seaborn provides several approaches to visualizing relational plots, such as scatter and line.

```python
# Scatter
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker', style='time', size='size', kind='scatter')
```

```python
# Line
sns.relplot(data=fmri, x='timepoint', y='signal', hue='region', style='event', kind='line')
```

# Distribution Plots
Seaborn provides several approaches to visualizing distribution plots, such as histogram, kde, ecdf as well as Jointplot and Pairplot.

### Univariate Distributions

```python
# Histogram
sns.displot(penguins, x='flipper_length_mm', hue='species', kind='hist')
```

```python
# Kernel density estimation
sns.displot(penguins, x='flipper_length_mm', hue='species', kind='kde')
```

```python
# Empirical cumulative distribution
sns.displot(penguins, x='flipper_length_mm', hue='species', kind='ecdf')
```

### Bivariate Distributions

```python
# Bivariate histogram
sns.displot(penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', kind='hist')
```

```python
# Bivariate kernel density estimation
sns.displot(penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', kind='kde')
```

```python
# Jointplot
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
```

```python
# Pairplot
sns.pairplot(penguins, hue='species')
```

# Categorical Plots
Seaborn provides several approaches to visualizing categorical plots, such as swarm, box, violin, count, bar, point, etc.

```python
# Swarm
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='swarm')
```

```python
# Box
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='box')
```

```python
# Violin
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='violin', split=True)
```

```python
# Count
sns.catplot(data=titanic, x='sex', hue='class', kind='count')
```

```python
# Bar
sns.catplot(data=titanic, x='sex', y='survived', hue='class', kind='bar')
```

```python
# Point
sns.catplot(data=titanic, x='class', y='survived', hue='sex', kind='point')
```

# Matrix Plots
Seaborn provides several approaches to visualizing matrix plots, such as headmap.

```python
# Pivot
flights = flights_long.pivot('month', 'year', 'passengers')
flights
```

```python
# Heatmap
sns.heatmap(flights, annot=True, fmt='d', linewidths=.5)
```

# Multi Plots
Seaborn provides several approaches to visualizing multiple plots, such as row and column parameters as well as FacetGrid and PairGrid.

```python
# Column
sns.lmplot(data=tips, x='total_bill', y='tip', col='time', hue='smoker')
```

```python
# Row
sns.lmplot(data=tips, x='total_bill', y='tip', row='time', hue='smoker')
```

```python
# FacetGrid
g = sns.FacetGrid(tips, col='sex', hue='smoker')
g.map(sns.scatterplot, 'total_bill', 'tip', alpha=.7)
g.add_legend()
```

```python
#PairGrid
g = sns.PairGrid(penguins, hue='species', corner=True)
g.map_lower(sns.kdeplot, hue=None, levels=5, color='.2')
g.map_lower(sns.scatterplot, marker='+')
g.map_diag(sns.histplot, element='step', linewidth=0, kde=True)
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61, .6))
```
